```python
import matplotlib.pyplot as plt
year=[2012, 2013, 2014, 2015, 2016, 2017]
salary=[12, 13, 14, 17, 19, 20]
population=[100, 120, 180, 250, 300, 370]

```

```python
plt.scatter(year, salary, s=population, c='r', alpha=0.5)
plt.title("Bubble Chart")
plt.ylabel("salary")
plt.xlabel("year")
plt.grid()
plt.text(year[0]+0.1, salary[0]+0.1, population[0])
plt.text(year[3]+0.1, salary[3]+0.1, population[3])
plt.show()
```

```python
# different color
import numpy as np
c=np.arange(len(year))
c[1]=c[0]
plt.scatter(year, salary, s=population, c=c, alpha=1, edgecolor='blue', marker='^')
plt.title("bubble chart")
plt.show()
```

```python
# pie chart
labels=["A","B","C","D"]
sizes=[3,4,6,2]
colors=["brown","blue","red","pink"]
explode=[0.1,0.2,0,0]
plt.title("distributing classes on basis of percentage")
plt.pie(sizes,colors=colors,explode=explode,labels=labels,autopct="%2.f%%", startangle=100, counterclock=False)
plt.show()
```
