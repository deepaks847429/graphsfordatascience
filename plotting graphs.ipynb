```python
import numpy as np
x=np.arange(0,5,0.1)
y=x**3
y2=x**2
plt.plot(x,y, color="black", linewidth=5, label="x^3")
plt.plot(x,y2, color="blue", linewidth=5,label="x^2")
plt.ylabel("y")
plt.xlabel("x")
plt.grid()
plt.text(2,80,"text", fontsize=18)
plt.axis([0,10,0,200])
plt.legend()
plt.plot(x,y)
```
